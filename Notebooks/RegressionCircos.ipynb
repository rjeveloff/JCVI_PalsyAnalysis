```python
# Import Libraries
import pandas as pd
import scipy.stats as stats
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from matplotlib import pyplot as plt
```

```python
chord_data = pd.DataFrame(columns=['Target','Source','Volume'])
data = pd.read_csv('../Datasets/clean.csv').set_index('ID')

gender_dict = {'F':1,'M':2}
data['GENDER'] = data['GENDER'].apply(lambda x: gender_dict[x])
```

```python
matrix = []
for src in data.columns:
    row = []
    for tgt in data.columns:
        if src == tgt:
            row.append(0)
            continue
        slope, intercept, r_value, p_value, std_err = stats.linregress(data[src],data[tgt])
        if np.isnan(r_value):
            row.append(0)
            continue
        r2 = r_value ** 2
        chord_data.loc[chord_data.shape[0]] = [src,tgt,r2]
        row.append(r2)
    matrix.append(row)
```
